```python
import pandas as pd

populations_df = pd.read_csv("Countries and population.csv")
gdp_df = pd.read_csv("Countries and GDP.csv")

populations_df.rename(columns={"Country/Territory": "Country"}, inplace=True)

merged_df = pd.merge(populations_df, gdp_df, on="Country", how="outer", indicator=True)

no_match_left = merged_df[merged_df['_merge'] == 'left_only']
no_match_right = merged_df[merged_df['_merge'] == 'right_only']

print("Record senza match nel file 'Countries and populations.csv':")
print(no_match_left["Country"].tolist())

print("Record senza match nel file 'Countries and GDP.csv':")
print(no_match_right["Country"].tolist())
```
