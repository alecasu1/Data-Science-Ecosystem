```python
# prompt: # TASK 3: Calculate the distances between a launch site to its proximities

# Assuming you have spacex_df and necessary libraries imported as in the previous code

# Function to calculate distance between two coordinates
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example usage (replace with your actual proximity coordinates)
launch_site_lat = spacex_df['Lat'].iloc[0]
launch_site_lon = spacex_df['Long'].iloc[0]
proximity_lat = 29.559684888503615  # Example proximity latitude
proximity_lon = -95.0830971930759 # Example proximity longitude

distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)
print(f"Distance to proximity: {distance} km")


# To calculate distances for all launch sites and their proximities you would need a list of proximities
# and iterate through them, calculating the distance for each launch site and each proximity

# Example for multiple proximities:
proximities = [
  [29.559684888503615, -95.0830971930759], # NASA
  [34.635, -120.623], # Example proximity 2
]

for lat, lon in proximities:
    distance = calculate_distance(launch_site_lat, launch_site_lon, lat, lon)
    print(f"Distance to proximity ({lat}, {lon}): {distance} km")
```
